```python
import numpy as np
import random as rd

class LinearRegression():

    def __init__(self, lr=0.001):
        self.a = rd.randint(-10, 10)
        self.b = rd.randint(-10, 10)
        self.lr = lr


    def f(self, x):
        return (self.a*x) + self.b


    def dot(self, x, y):
        if len(x) != len(y):
            return 'Erreur'
        else:
            prod=0
            for i in range(len(x)):
                prod+= x[i]*y[i]
            return prod


    def sum_(self, x):
        return sum(x)


    def loss(self, x, y):
        if len(x) != len(y):
            return 'Erreur'
        
        else:
            loss=0
            for i in range(len(x)):
                loss+=(self.f(x[i])-y[i])**2
            return loss

    def  grad(self, x, y):
        if len(x) != len(y):
            return 'Erreur'
        else:
            delta_a = [x[i]*(y[i] - self.f(x[i])) for i in range(len(x))]
            delta_b = [y[i] - self.f(x[i]) for i in range(len(x))]
            a = -(0.5*len(x))*(self.sum_(delta_a))
            b = -(0.5*len(x))*(self.sum_(delta_b)) 
            return a, b

    
    def fit (self, x, y):
        count = 0
        loss = 10
        print(self.a, self.b)
        while (abs(loss) > 0.001): # and (count<10000):
            count+=1
            loss = self.loss(x, y)
            print(loss, count)
            grad_a, grad_b = self.grad(x, y)
            print(grad_a, grad_b)
            self.a = self.a - self.lr*grad_a
            self.b = self.b - self.lr*grad_b
        return (self.a, self.b)
```
